```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# PREPROCESAMIENTO DE LOS DATOS Y ETL

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

Selección de columnas

```python
columns = [
    "EmployeeNumber",
    "Age",
    "Attrition",
    "Department",
    "JobRole",
    "JobLevel",
    "JobSatisfaction",
    "MonthlyIncome",
    "YearsAtCompany",
    "TotalWorkingYears",
    "TrainingTimesLastYear",
    "OverTime",
    "WorkLifeBalance"
]
df = df[columns]
df
```

Limpieza de Datos

```python
# OVERVIEW

df.info()
```

```python
#NULOS

for i in df.columns:
    print(f"Valores Null en {i}: {df[i].isnull().sum()}")
```

```python
#ÚNICOS

for i in df.columns:
    print(f"Valores Null en {i}: {df[i].unique()}")
```

```python
# DUPLICADOS

df = df.drop_duplicates(subset=["EmployeeNumber"])
df
```

```python
# CATEGÓRICOS

df["Attrition"] = np.where(df["Attrition"] == "Yes", 1, 0)
df
```

```python
# VALORES POSITIVOS

df = df[df["MonthlyIncome"] >= 0]
df = df[df["YearsAtCompany"] >= 0]
df
```

<1 año - Onboarding

1-3 años - Carrera

3-5 años - Estabilidad

+5 años - Clave

```python
# ADD COLUMNA ETIQUETAS

def tenure_group(years):
    if years < 1:
        return "Onboarding"
    elif years < 3:
        return "Carrera"
    elif years < 5:
        return "Estable"
    else:
        return "Clave"

df["tenure_group"] = df["YearsAtCompany"].apply(tenure_group)

df
```

# EDA - Análisis Exploratorio

Attrition = 1 - Deserción

Attrition = 0 - Activo

Tasa de rotación general

```python
total_employees = len(df) # dado que comprobamos que el ID es único
total_attrition = df["Attrition"].sum() # Suma de 'Trues'
turnover_rate = total_attrition / total_employees
print(f"El turnover rate total es de: {(turnover_rate*100).round(2)}%")
```

Tasa de rotación por Departament

```python
turnover_department = (df.groupby("Department")["Attrition"].mean().sort_values(ascending=False)) #Agrupamos por Department y de cada grupo obtenemos 'attrition', obtenemos el promedio y ordenamos los valores
(turnover_department*100).round(2)
```

Tasa de rotación por Job Role

```python
turnover_jobrole = (df.groupby("JobRole")["Attrition"].mean().sort_values(ascending=False)) 
(turnover_jobrole*100).round(2)
```

Edad Vs Deserción

```python
df.groupby("Attrition")['Age'].mean().round() #Obtenemos la edad promedio de los que desertaron
```

Rangos de Edad

```python
def age_group(age):
    if age < 30:
        return 'Menor a 30'
    elif age < 40:
        return '30 a 39'
    elif age < 50:
        return '40 a 49'
    else:
        return '50 o más'

df["age_group"] = df["Age"].apply(age_group)

turnover_age = (df.groupby('age_group')['Attrition'].mean().sort_values(ascending=False))
print((turnover_age*100).round(2))
```

Correlación

```python
#MATRIZ DE CORRELACIÓN
numeric_columns = [
    'Age',
    'MonthlyIncome',
    'TotalWorkingYears',
    'YearsAtCompany',
    'Attrition'
]
correlation_matrix = df[numeric_columns].corr()
correlation_matrix
```

```python
# HEATMAP
corr = df.select_dtypes(include='number').corr() # Tomamos solo las variables numéricas

plt.figure(figsize=(10, 8))
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title('Matriz de Correlación')
plt.tight_layout()
plt.show()
```

Gráfica turnover vs grupo de edad

```python
turnover_age_pct = turnover_age * 100

plt.figure(figsize=(8, 5))

bars = plt.bar(
    turnover_age_pct.index,
    turnover_age_pct.values,
    color=['#d73027', '#fc8d59', '#91bfdb', '#4575b4']
)

plt.title('Turnover Rate por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Turnover Rate (%)')

# Agregamos las etiquetas en cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        f'{height:.1f}%',
        ha='center',
        va='bottom'
    )

plt.ylim(0, max(turnover_age_pct.values) + 5)
plt.tight_layout()
plt.show()
```

COSTOS

```python
#COSTO DE ROTACIÓN - REEMPLAZO DE EMPLEADO
df["estimated_cost"] = df["MonthlyIncome"] * 3 #Creamos la columna con los posibles costos de reemplazar a cada empleado 
```

```python
#COSTO DE LOS QUE SE FUERON
attrition_df = df[df['Attrition'] == 1] #Nos enfocamos solo en los que se fueron
#TOTAL
total_turnover_cost = attrition_df['estimated_cost'].sum()
print(f"El costo total de la rotación es de: ${total_turnover_cost}")
```

```python
# COSTO PROMEDIO POR EMPLEADO
avg_turnover_cost = attrition_df['estimated_cost'].mean()
print(f"El costo promedio por empleado es de: ${avg_turnover_cost.round(2)}")
```

```python
#COSTO POR DEPARTAMENTO
cost_by_department = (attrition_df.groupby('Department')['estimated_cost'].sum().sort_values(ascending=False))
#COSTO POR JOB ROLE
cost_by_jobrole = (attrition_df.groupby('JobRole')['estimated_cost'].sum().sort_values(ascending=False))
print(f"El costo promedio por departamento es de: ${cost_by_department}")
print(f"El costo promedio por rol del empleado es de: ${cost_by_jobrole}")
```

GRÁFICA COSTOS

```python
#DEFINIMOS EL COSTO POR DEPARTAMENTO
cost_dept = (df[df['Attrition'] == 1].groupby('Department')['estimated_cost'].sum())

# GRÁFICO POR DEPARTAMENTO
plt.figure(figsize=(6, 6))
cost_dept.plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.title('Costo de rotación por departamento')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#DEFINIMOS EL COSTO POR ROLE
cost_role = (df[df['Attrition'] == 1].groupby('JobRole')['estimated_cost'].sum().sort_values(ascending=False)) #podemos añadir .head(5) para un top 5, en caso de tener muchos roles

# GRÁFICO POR ROLE
plt.figure(figsize=(6, 6))
cost_role.plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.title('Distribución del costo de rotación por roles clave')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
